```python
import cv2
from google.colab.patches import cv2_imshow

print("Experiment 32 - Image Pyramid")

img = cv2.imread("sample 8.jpg")

if img is None:
    print("Upload sample 8.jpg")
else:
    print("INPUT IMAGE:")
    cv2_imshow(img)

    smaller = cv2.pyrDown(img)

    print("OUTPUT IMAGE:")
    cv2_imshow(smaller)
```
