```python
import cv2
from google.colab.patches import cv2_imshow

print("Experiment 29 - Bitwise AND")

img1 = cv2.imread("sample 9.jpg")
img2 = cv2.imread("sample 10.jpg")

if img1 is None or img2 is None:
    print("Upload sample29_1.jpg and sample29_2.jpg")
else:
    print("INPUT IMAGE 1:")
    cv2_imshow(img1)

    print("INPUT IMAGE 2:")
    cv2_imshow(img2)

    # Check if images have the same shape and resize if necessary
    if img1.shape != img2.shape:
        print(f"Image shapes mismatch. Resizing img2 from {img2.shape} to {img1.shape}.")
        img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

    result = cv2.bitwise_and(img1, img2)

    print("OUTPUT IMAGE:")
    cv2_imshow(result)
```
