```python
import cv2
from google.colab.patches import cv2_imshow

print("Experiment 27 - Laplacian Edge Detection")

img = cv2.imread("sample 7.png")

if img is None:
    print("Upload sample 7.png")
else:
    print("INPUT IMAGE:")
    cv2_imshow(img)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    lap = cv2.Laplacian(gray, cv2.CV_64F)

    print("OUTPUT IMAGE:")
    cv2_imshow(lap)
```
