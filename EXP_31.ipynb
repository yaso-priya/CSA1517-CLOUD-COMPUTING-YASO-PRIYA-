```python
import cv2
from google.colab.patches import cv2_imshow

print("Experiment 31 - Image Blending")

img1 = cv2.imread("sample 8.jpg")
img2 = cv2.imread("sample 7.jpg")

if img1 is None or img2 is None:
    print("Upload both JPG images")
else:
    # Resize img2 to match the dimensions of img1
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

    print("INPUT IMAGE 1:")
    cv2_imshow(img1)

    print("INPUT IMAGE 2:")
    cv2_imshow(img2)

    blended = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)

    print("OUTPUT IMAGE:")
    cv2_imshow(blended)
```
