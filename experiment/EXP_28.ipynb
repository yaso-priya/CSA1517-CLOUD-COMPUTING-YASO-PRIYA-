```python
import cv2
from google.colab.patches import cv2_imshow

print("Experiment 28 - Sobel Edge Detection")

img = cv2.imread("sample 8.png")

if img is None:
    print("Upload sample 8.png")
else:
    print("INPUT IMAGE:")
    cv2_imshow(img)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    sobel = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)

    print("OUTPUT IMAGE:")
    cv2_imshow(sobel)
```
